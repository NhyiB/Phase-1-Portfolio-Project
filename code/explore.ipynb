### Importing the necessary packages 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Loading the data into a dataframe
#Note we had to specify the semicolon after reading in the file because the dataset is seperated by that instaed of a comma

df = pd.read_csv("/Users/sa03/Phase-1-Portfolio-Project/data/dataset_Facebook.csv" , sep = ";")
```

### Checking the quality of our data

```python
#Finding the sum of our missing values in our dataset

df.isnull().sum()
```

```python
# We see that the likes and shares column have a few null values lets check the percentage to the whole data to decide what to do with them

(df[['like','share']].isnull().sum()/len(df)) * 100
```

```python
#We see from above that the percentages are very small and hence dropping it will niot affect our analysis so lets drop the null values 

df = df.dropna(subset=['like','share'])
```

```python
# Checking to see if we dropped them successfully 

df[['like','share']].isnull().sum()
```

```python
# We can see that our dataset has no duplicates so that is good 
df.duplicated()
```

```python
df.info()
```

### Converting the object data type to numeric

```python
#We can see from our df.info above that the data type for Type is object instead of nueric and we need to change that so we can run our final analysis successfully

df['Type'].unique()
```

```python
# Now we do the converting

df = pd.get_dummies(df, columns = ['Type'], drop_first= True)
```

```python
df.head()
```

```python
# Lets explore our target variable

sns.histplot(df["Total Interactions"],kde = True)
plt.title('Distribution of Total interactions')
plt.show()

df["Total Interactions"].skew()
```

```python
sns.boxplot(y=df["Total Interactions"])
plt.title('Distribution of Total interactions')
plt.show()
```

 We see from above that the total interactions column has a positive skew. This means most of our dataset is between 0 and 1000 interactions but there are a few outliers that strech our data all the way. We would take a look at thus later by performing log transformations on it before modeling our data.

```python
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns
numeric_cols

df[numeric_cols].hist(figsize=(12,10), bins=20)
plt.tight_layout()
plt.show()
```

```python
#Ploting our heatmap to find the correlation of the various variables with the target variable.
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns

corr = df[numeric_cols].corr()
plt.figure(figsize=(12,12))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()

corr["Total Interactions"].sort_values(ascending=False)
```
